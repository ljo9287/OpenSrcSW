```python
import matplotlib.pyplot as plt
import numpy as np
import csv 

f = open('age.csv')
data = csv.reader(f)
next(data)
data = list(data)
home = []
min = 9999999999
max = 0
result_name_min = ''
result_name_max = ''
result_min = 0
result_max = 0
name = '서울특별시 광진구'

for row in data:
    for i in range(2,105):
        row[i] = int(row[i].replace(',', ''))
    if name in row[1]:
        home = np.array(row[4:], dtype=int)

for row in data:
    away_min = np.array(row[4:], dtype=int)
    s = np.sum((home-away_min)**2)
    
    if s < min and name not in row[1]:
        min = s 
        result_name_min = row[1]
        result_min = away_min

for row in data:
    away_max = np.array(row[4:], dtype=int)
    s = np.sum((home-away_max)**2)
  
    if s > max and name not in row[1]:
        max = s 
        result_name_max = row[1]
        result_max = away_max

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name + '와 가장 비슷한 지역과 다른 지역')
plt.plot(home, label = name)
plt.plot(result_min, label=result_name_min + str(np.std(result_min)))
plt.plot(result_max, label=result_name_max + str(np.std(result_max)))
plt.legend()
plt.show()

```
