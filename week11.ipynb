```python
def copy(fname, fcopy):
    f1 = open(fname, 'r')
    f2 = open(fcopy, 'w')

    data = f1.read()
    f2.write(data)

    print(data)

    f1.close()
    f2.close()
    
if __name__ == '__main__':
    fname = input('file1')
    fcopy = input('file2')

    copy(fname, fcopy)
```

```python
def wc(fname):
    dic = {}
    f = open(fname, 'r')
    print('lines:',len(f.readlines()))
    f.close()

    f = open(fname, 'r')

    for line in f:
        line = line.replace('\n', '')
        words = line.split()
        for word in words:
            if word in dic:
                dic[word] += 1
            else:
                dic[word] = 1
    for word in dic:
        print(word, dic[word])

    f.close()

if __name__ == '__main__':
    fname = input('fname: ')
    wc(fname)
    
```

```python
class Calc:

    def sum(self, a, b):
        result = a + b
        print('{0} + {1} = {2}'.format(a, b, result))
        
    def sub(self, a, b):
        result = a - b
        print('{0} - {1} = {2}'.format(a, b, result))
        
    def multi(self, a, b):
        result = a * b
        print('{0} * {1} = {2}'.format(a, b, result))

    def divi(self, a, b):
        result = a / b
        print('{0} / {1} = {2}'.format(a, b, result))


if __name__ == '__main__':
    print('calc_ex')
    a = Calc()
    
    print(a.sum(1,2))
    print(a.sub(1,2))
    print(a.multi(1,2))
    print(a.divi(1,2))
```

```python
a=input()
stack=[]
res=''
ans = []

#중위식 -> 후위식
for x in a:
    if x.isdecimal():
        res += x
    else:
        
        if x=='*' or x=='/':
            while stack and (stack[-1]=='*' or stack[-1]=='/'):
                res+=stack.pop()
            stack.append(x)
        elif x=='+' or x=='-':
            while stack:
                res+=stack.pop()
            stack.append(x)
       
while stack:
    res+=stack.pop()
print(res)

operations = {
            '*': lambda x, y: y * x,
            '/': lambda x, y: y / x,
            '+': lambda x, y: y + x,
            '-': lambda x, y: y - x
            }


for x in res:
    if x not in operations:
        stack.append(int(x))

    elif x == '+':
        n1 = stack.pop()
        n2 = stack.pop()
        stack.append(operations[x](n1,n2))

    elif x == '-':
        n1 = stack.pop()
        n2 = stack.pop()
        stack.append(operations[x](n1,n2))
    
    elif x == '*':
        n1 = stack.pop()
        n2 = stack.pop()
        stack.append(operations[x](n1,n2))

    elif x == '/':
        n1 = stack.pop()
        n2 = stack.pop()
        stack.append(operations[x](n1,n2))
        
for i in stack:
    ans.append(i)
print(ans)
```
