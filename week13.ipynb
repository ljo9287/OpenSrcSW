```python
import matplotlib.pyplot as plt
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result_top = []
result_low = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '24':
            result_top.append(float(row[-1]))
plt.plot(result_top, label = 'top')
plt.legend()

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result_top = []
result_low = [] 
for row in data:
    if row[-2]!= '':
        if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '24':
            result_low.append(float(row[-2]))
try:
    plt.plot(result_low, label = 'low')
except Exception as e:
    print(e)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

from numpy import short
f = open('seoul.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

plt.hist(aug, bins=100, color='r', label='aug')
plt.hist(jan, bins=100, color='g', label='jan')
plt.legend()
plt.show() 
```
