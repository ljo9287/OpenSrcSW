```python

members = {}
num = 0

while num < 10:
    menu = input('회원정보 추가(a), 검색(b)')
    if menu=='a':
        name = input('이름: ')
        phone = int(input('전화번호: '))
        members[name] = phone
        print('----------')
        print('{0:s} 입력 완료'.format(name))
        print('{0:s}: {1:d}'.format(name, phone))
        #print(f'{name}:', phone)
        print('----------')
        num = num + 1

    elif menu=='b':
        name = input('이름: ')
        phone = members.get(name)
        
        if name not in members:
            print('전화번호부 목록에 존재하지 않습니다.') 
        
        else:
            print('----------')
            print('{0:s}: {1:d}'.format(name, phone))
            print('----------')
```
